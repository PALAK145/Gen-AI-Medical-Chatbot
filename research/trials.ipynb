```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
def load_pdf_file(data):
    loader = DirectoryLoader(
        data,
        glob="**/*.pdf",          # recursive
        loader_cls=PyPDFLoader,
        show_progress=True,       # needs ipywidgets/tqdm installed
        use_multithreading=True,  # parallelize across files
        silent_errors=True,       # don't hang/crash on one bad PDF, just skip + log
    )
    return loader.load()
```

```python
import os
path = 'C:\Generative AI\Gen-AI-Medical-Chatbot\Data\The-Gale-Encyclopedia-of-Medicine-3rd-Edition-staibabussalamsula.ac_.id_.pdf' 

print("Size (MB):", os.path.getsize(path) / 1e6)

with open(path, 'rb') as f:
    header = f.read(5)
    print("Header:", header)  # should be b'%PDF-'
```

```python
import pypdf
reader = pypdf.PdfReader(path)
print("Pages:", len(reader.pages))
print("Encrypted:", reader.is_encrypted)
```

```python
import time
reader = pypdf.PdfReader(path)
for i, page in enumerate(reader.pages):
    start = time.time()
    text = page.extract_text()
    print(f"Page {i}: {time.time()-start:.2f}s, {len(text or '')} chars")
```
